```python
import pandas as pd
import requests
import json
```

Set the base url, and formatted url for barley only data in the United States. Retrieve seperate attributes description and measurement descriptions.

```python
base_url = 'api.fas.usda.gov'
attribute_url = 'https://api.fas.usda.gov/api/psd/commodityAttributes'
measurements_url = 'https://api.fas.usda.gov/api/psd/unitsOfMeasure'
barley_url = 'https://api.fas.usda.gov/api/psd/commodity/0430000/country/us/year'
```

Retrieve my secrete key for the API

```python
with open("USDAkey.json", "r", encoding="utf-8") as file:
    api_key = json.load(file)
    secret_api_key = api_key.get("X-Api-Key", "")
```

Function to easily retrive the different endpoints.

```python
def api_call(url):
    headers = {
        "X-Api-Key": secret_api_key
    
    }
    response = requests.get(url, headers=headers)
    response.raise_for_status()

    return response.json()

```

```python
attributes = api_call(attribute_url)
measurements = api_call(measurements_url)
```

Select data for a defined number of years.

```python
years = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]
data_list = []

for year in years:
    year_url = f"{barley_url}/{year}"  
    try:
        data = api_call(year_url)
        for entry in data:  
            entry['year'] = year  
            data_list.append(entry)  
        print(f"Data retrieved for {year}")
    except requests.exceptions.RequestException as e:
        print(f"Error fetching data for {year}: {e}")


```

1. Convert retrieved list into a Pandas dataframe

```python
df = pd.DataFrame(data_list)
df.head()

```

2 and 3. Create column the identifies the attributes name and create a column that identifies the units of measure.

```python
attribute_dict = {item['attributeId']: item['attributeName'] for item in attributes}
measurement_dict = {item['unitId']: item['unitDescription'] for item in measurements}

df['attributeName'] = df['attributeId'].map(attribute_dict)
df['units'] = df['unitId'].map(measurement_dict)

df.head()


```

4. find duplicates

```python
df.duplicated().sum()
```

5. Detect outliers

```python
from matplotlib import pyplot as plt
import seaborn as sns
plt.figure(figsize=(12, 6))


sns.boxplot(data= df, x= 'attributeName', y = 'value')
plt.xticks(rotation=90)
plt.title("Box Plot of Value by Attribute Name")


```

There are some outliers in Beginning Stock, Total Supply, Total Year Exports and Total Distribution. I am going to leave these outliers as they may be indicators of climate affects.

This data was pretty clean, I needed to use two endpoints because the data only had ID numbers. The second endpoint contained the descriptors. The first thing that I did was convert the json files to pandas dataframes. I then created a column that contained the attribute names from the second endpoint and then created a column that indicates the units of measure from the second endpoint. Their doesn't appear to be any ethical concerns with this data, since I needed to register on the site for a security key and nothing was force downloaded.

Create new database

```python
import sqlite3
import pandas as pd

# filename to form database
file = "DSC540.db"

try:
  conn = sqlite3.connect(file)
  print("Database DSC540.db formed.")
except:
  print("Database DSC540.db not formed.")

c = conn.cursor()
```

Read in the three seperate files.

```python
flat_df = pd.read_csv('BarleyData.csv')
web_df = pd.read_csv('BarleyWeb.csv')
api_df = pd.read_csv('BarleyAPI.csv')
```

Move the dataframes as tables in the database.

```python
flat_df.to_sql('Barleyflat',con = conn, if_exists= 'replace', index=False)
web_df.to_sql('Barleyweb', con=conn, if_exists= 'replace', index=False)
api_df.to_sql('Barleyapi', con=conn, if_exists= 'replace', index=False)
```

Verify the tables.

```python
c.execute('SELECT * FROM Barleyflat')
data = c.fetchall()
data
```

Join the three database tables  into one table on the database.

```python
c.execute("""
    CREATE TABLE IF NOT EXISTS BarleyCombined AS
    SELECT 
        a.*, f.*, w.*
    FROM BarleyAPI a
    JOIN Barleyflat f ON a.calendarYear = f.Year_ID
    JOIN Barleyweb w ON a.calendarYear = w."Market Year"
""")
c.close()
conn.close()
```

Read in the combined table as a Pandas dataframe.

```python
conn = sqlite3.connect('DSC540.db')


df = pd.read_sql_query("SELECT * FROM BarleyCombined", conn)
df.head()
```

Convert the measurements into something that is normal for farmers, bushels per acre. Then draw a line plot indicating bushels per acre over time.

```python
METRIC_TON_TO_KG = 1000
HECTARE_TO_ACRE = 2.471
BARLEY_BUSHEL_WEIGHT_KG = 21.77


df_filtered = df[df['attributeName'] == 'Yield']

df_grouped = df_filtered.groupby('marketYear')['value'].mean().reset_index()

sns.lineplot(data=df_grouped, x='marketYear', y='value')

plt.xlabel('Year')
plt.ylabel('Total Yield')
plt.title('Annual Bushel per Acre of Barley Yield')

```

I want to compare bushels per acre vs acres planted. Convert area harvested to bushels per acre.

```python
df_filtered_acres = df[df['attributeName'] == 'Area Harvested']
df_filtered_acres['value'] = (df_filtered_acres['value'] * METRIC_TON_TO_KG / BARLEY_BUSHEL_WEIGHT_KG) / HECTARE_TO_ACRE
df_merged = pd.merge(df_grouped, df_filtered_acres, on='marketYear')

```

Create a scatter plot with regression showing the correlation.

```python
sns.regplot(data=df_merged, x='value_y', y='value_x')

plt.xlabel('Acres Planted')
plt.ylabel('Bushels per Acre')
plt.title('Acres Planted vs Yield (Bushels per Acre)')

```
